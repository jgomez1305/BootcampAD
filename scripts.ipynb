## Limpieza de Datos
En esta sección, realizamos la limpieza de datos para asegurarnos de que esté listo para el análisis. Esto incluye la identificación y manejo de valores nulos, así como la revisión y eliminación de datos duplicados .


### Valores Nulos

1. **Identificación de Valores Nulos**: Verificamos si hay valores nulos.

2. **Manejo de Valores Nulos**: Eliminamos las filas con valores nulos para garantizar la integridad de los datos.

```python
# Identificar si hay valores nulos o vacíos en el DataFrame
check_null = df.isnull().any().any()

if check_null:
    print("Existen valores nulos o vacíos en los datos.")

    # Mostrar filas con valores nulos o vacíos
    rows_with_null = df[df.isnull().any(axis=1)]
    print("Filas con valores nulos o vacíos:")
    print(rows_with_null)

    # Eliminar filas con valores nulos o vacíos
    df = df.dropna()
    print("Los valores nulos o vacíos fueron eliminados.")
else:
    print("No hay valores nulos o vacíos en los datos.")
```

### Datos Duplicados

1. **Revisión de Datos Duplicados**: Buscamos datos duplicados.

2. **Eliminación de Datos Duplicados**: Eliminamos las filas duplicadas para evitar redundancias en los datos.


```python
# Revision de datos duplicados
check_dup = df.duplicated().any()
print("¿Existen valores duplicados en los datos?",check_dup)

duplicateRowsDF = df[df.duplicated()]
print(duplicateRowsDF)#Se utiliza para saber cuales son los valores duplicados

if check_dup:
  df=df.drop_duplicates()
  print("Los valores duplicados fueron eliminados.")
else:
  print("No hay valores duplicados en los datos.")
```

#Identificacion de los valores para poder traducir

```python
#Traduccion de datos

tra_genero={'Female':'Mujer','Male':'Hombre'}
tra_sobrepeso={'yes':'Si','no':'No'}
tra_favc={'yes':'Si','no':'No'}
tra_Obesidad={'Insufficient_Weight': 'PesoInsuficiente','Normal_Weight': 'PesoNormal','Overweight_Level_I': 'SobrepesoNivel1','Overweight_Level_II': 'SobrepesoNivel2',
              'Obesity_Type_I':'ObesidadTipo1','Obesity_Type_II': 'ObesidadTipo2','Obesity_Type_III':'ObesidadTipo3'}
tra_transporte={'Public_Transportation':'TransportePublico', 'Walking':'Caminar', 'Automobile':'Automovil', 'Motorbike':'Moto', 'Bike':'Cicla'}
tra_alcohol={'no':'No', 'Sometimes':'Aveces', 'Frequently':'Frecuentemente', 'Always':'Siempre'}

df['Genero'] = df['Genero'].replace(tra_genero)
df['FamiliaConSobrepeso'] = df['FamiliaConSobrepeso'].replace(tra_sobrepeso)
df['FAVC'] = df['FAVC'].replace(tra_favc)
df['TObesidad'] = df['TObesidad'].replace(tra_Obesidad)
df['TransporteUtilizado'] = df['TransporteUtilizado'].replace(tra_transporte)
df['CALC'] = df['CALC'].replace(tra_alcohol)
```


## Reinicio de Datos de Columnas
En esta sección, se eliminara todos los datos existentes de dos columnas y la posterior adición de nuevos valores aleatorios.

```python
#Para generar datos aletorios necesitamos la siguiente libreria
import random
```

```python
# Eliminamos todos los datos de la columna 'FCVC'
df['FCVC'] = ''

# Probabilidad de obtener "No"
probabilidad_no = 0.7  # Por ejemplo, 70% de probabilidad de obtener "No"

# Cambiar los valores en la columna 'FCVC' según la probabilidad especificada
for i in range(len(df)):
    if random.random() < probabilidad_no:
        df.at[i, 'FCVC'] = "No"
    else:
        df.at[i, 'FCVC'] = "Sí"

# Mostrar el conteo de "Sí" y "No"
conteo = df['FCVC'].value_counts()
print(conteo)
```

```python
# Eliminar filas con valores vacíos en la columna 'FCVC'
df = df.drop(df[df['FCVC'] == ''].index)

# Mostrar el nuevo conteo de "Sí" y "No" después de la eliminación
conteo = df['FCVC'].value_counts()
print(conteo)
```

```python
# Eliminamos todos los datos de la columna 'CH2O'
df['CH2O'] = ''
# Probabilidad de obtener "RaraVez"
probabilidad_raravez = 0.6  # 60% de probabilidad de obtener "RaraVez"

# Cambiar los valores en la columna 'CH2O' según la probabilidad especificada
for i in range(len(df)):
    if random.random() < probabilidad_raravez:
        df.at[i, 'CH2O'] = "RaraVez"
    else:
        df.at[i, 'CH2O'] = "Siempre"

# Mostrar el conteo de "RaraVez" y "Siempre"
conteo = df['CH2O'].value_counts()
print(conteo)
```

## Diccionario de Datos

A continuación, se presenta un diccionario de datos que describe las columnas y sus respectivos valores en el conjunto de datos:

- **Genero**: Representa el género de la persona y puede tener los valores 'Mujer' o 'Hombre'.

- **Edad**: Indica la edad de la persona en años.

- **Altura**: Muestra la altura de la persona en centímetros.

- **Peso**: Indica el peso de la persona en kilogramos.

- **FamiliaConSobrepeso**: Describe si la persona tiene antecedentes familiares con sobrepeso y tiene valores de 'si' o 'no'.

- **TObesidad**: Describe la categoría de obesidad de la persona y tiene los siguientes valores: 'PesoInsuficiente,' 'PesoNormal,' 'SobrepesoNivel1,' 'SobrepesoNivel2,' 'ObesidadTipo1,' 'ObesidadTipo2,' y 'ObesidadTipo3'.

- **TransporteUtilizado**: Describe el transporte frecuentemente utilizado por la persona. Tiene como valor 'TransportePublico', 'Moto', 'Automovil', 'Caminar', 'Cicla.

- **Consumo frecuente de alimentos ricos en calorías (FAVC)**: Indica si la persona tiene el hábito de consumir alimentos con alta densidad calórica. tiene el valor 'si' si es frecuente o 'no' si no es el caso.

- **Frecuencia de consumo de verduras (FCVC)**: Describe la frecuencia con la que la persona consume verduras. tiene el valor 'si' si es frecuente o 'no' si no es el caso.

- **Consumo de agua diario (CH20)**: Describe la frecuencia con la que la persona consume agua dia. tiene el valor 'Siempre' si consume aunque sea un solo baso de agua por dia o 'Raravez' si consume agua de vez en cuando.

- **Consumo de alcohol (CALC)**: Describe la frecuencia con la que la persona consume alcohol. tiene el valor de 'No', 'A veces' , 'Frecuentemente' , 'Siempre'.

- **Frecuencia de actividad fisica(FAF)**: Describe la frecuencia con la que la persona hace ejercicio.

- **Frecuencia usando dispositivos tecnológicos(TUE)**: Describe la frecuencia con la que la persona hace uso de dispositivos tecnologicos.
