## Carga de Datos desde un Archivo CSV en GitHub

## Limpieza de Datos
En esta sección, realizamos la limpieza de datos para asegurarnos de que esté listo para el análisis. Esto incluye la identificación y manejo de valores nulos, así como la revisión y eliminación de datos duplicados.


### Valores Nulos

1. **Identificación de Valores Nulos**: Verificamos si hay valores nulos.

2. **Manejo de Valores Nulos**: Eliminamos las filas con valores nulos para garantizar la integridad de los datos.

```python
# Identificar si hay valores nulos o vacíos en el DataFrame
check_null = df.isnull().any().any()

if check_null:
    print("Existen valores nulos o vacíos en los datos.")

    # Mostrar filas con valores nulos o vacíos
    rows_with_null = df[df.isnull().any(axis=1)]
    print("Filas con valores nulos o vacíos:")
    print(rows_with_null)

    # Eliminar filas con valores nulos o vacíos
    df = df.dropna()
    print("Los valores nulos o vacíos fueron eliminados.")
else:
    print("No hay valores nulos o vacíos en los datos.")
```

### Datos Duplicados

1. **Revisión de Datos Duplicados**: Buscamos datos duplicados.

2. **Eliminación de Datos Duplicados**: Eliminamos las filas duplicadas para evitar redundancias en los datos.


```python
# Revision de datos duplicados
check_dup = df.duplicated().any()
print("¿Existen valores duplicados en los datos?",check_dup)

duplicateRowsDF = df[df.duplicated()]
print(duplicateRowsDF)#Se utiliza para saber cuales son los valores duplicados

if check_dup:
  df=df.drop_duplicates()
  print("Los valores duplicados fueron eliminados.")
else:
  print("No hay valores duplicados en los datos.")
```

##Eliminación de Columnas que no van a ser Utilizadas
En esta sección, se identificarán y eliminarán las columnas que no se consideran esenciales para el análisis, lo que simplificará los datos y mejorará su calidad.


```python
#Columnas eliminadas
df = df.drop(columns=['CAEC', 'CALC','SCC','TUE','SMOKE','MTRANS','NCP'])
```

## Traducción de Datos

En esta sección, realizamos la traducción de datos de las columnas relevantes de nuestro conjunto de datos de origen en inglés a su equivalente en español. Esto es esencial para asegurarnos de que nuestros datos estén en el idioma deseado y sean más comprensibles.



```python
#Renombrar encabezado de columnas
new_col={ 'Gender':'Genero','Age':'Edad','Height':'Altura','Weight':'Peso','family_history_with_overweight':'familia-con-sobrepeso', 'NObeyesdad':'TObesidad'}
df=df.rename(columns=new_col)
```

```python
df.head()
```

```python
#Identificacion de los valores para poder traduccir
valores_unicos = df['TObesidad'].unique()
print(valores_unicos)
```

```python
#Traduccion de datos

tra_genero={'Female':'Mujer','Male':'Hombre'}
tra_sobrepeso={'yes':'si'}
tra_favc={'yes':'si'}
tra_Obesidad={'Insufficient_Weight': 'PesoInsuficiente','Normal_Weight': 'PesoNormal','Overweight_Level_I': 'Sobrepeso_Nivel1','Overweight_Level_II': 'Sobrepeso_Nivel2',
              'Obesity_Type_I':'Obesidad_Tipo1','Obesity_Type_II': 'Obesidad_Tipo2','Obesity_Type_III':'Obesidad_Tipo3'}

df['Genero'] = df['Genero'].replace(tra_genero)
df['familia-con-sobrepeso'] = df['familia-con-sobrepeso'].replace(tra_sobrepeso)
df['FAVC'] = df['FAVC'].replace(tra_favc)
df['TObesidad'] = df['TObesidad'].replace(tra_Obesidad)
```

## Diccionario de Datos

A continuación, se presenta un diccionario de datos que describe las columnas y sus respectivos valores en el conjunto de datos:

- **Genero**: Representa el género de la persona y puede tener los valores 'Mujer' o 'Hombre'.

- **Edad**: Indica la edad de la persona en años.

- **Altura**: Muestra la altura de la persona en centímetros.

- **Peso**: Indica el peso de la persona en kilogramos.

- **familia-con-sobrepeso**: Describe si la persona tiene antecedentes familiares con sobrepeso y puede tener el valor 'si' o 'no'.

- **TObesidad**: Describe la categoría de obesidad de la persona y puede tener los siguientes valores: 'PesoInsuficiente,' 'PesoNormal,' 'Sobrepeso_Nivel1,' 'Sobrepeso_Nivel2,' 'Obesidad_Tipo1,' 'Obesidad_Tipo2,' y 'Obesidad_Tipo3'.

- **Consumo frecuente de alimentos ricos en calorías (FAVC)**: Este atributo indica si la persona tiene el hábito de consumir alimentos con alta densidad calórica. Puede tener el valor 'si' si es frecuente o 'no' si no es el caso.

- **Frecuencia de consumo de verduras (FCVC)**: Describe la frecuencia con la que la persona consume verduras por semana. Este atributo nos proporciona información sobre su ingesta de verduras.

- **Consumo de agua diario (CH20)**: Muestra la cantidad de agua que la persona consume diariamente. El agua es esencial para la hidratación y la salud en general.

- **Consumo de alcohol (CALC)**: Aunque no se menciona en la descripción anterior, este atributo podría estar relacionado con los hábitos alimentarios de las personas y su consumo de alcohol.
